```python
%sql
USE CATALOG fintech_finpay;
USE SCHEMA gold;
```

```python
%sql
REFRESH MATERIALIZED VIEW fintech_finpay.gold.dim_merchant;
REFRESH MATERIALIZED VIEW fintech_finpay.gold.dim_user;
REFRESH MATERIALIZED VIEW fintech_finpay.gold.dim_channel;
REFRESH MATERIALIZED VIEW fintech_finpay.gold.dim_date;
```

```python
%sql
REFRESH MATERIALIZED VIEW fintech_finpay.gold.fact_transactions;
```

```python
%sql
SELECT 'dim_merchant' AS object_name, COUNT(*) AS records FROM fintech_finpay.gold.dim_merchant
UNION ALL
SELECT 'dim_user' AS object_name, COUNT(*) AS records FROM fintech_finpay.gold.dim_user
UNION ALL
SELECT 'dim_channel' AS object_name, COUNT(*) AS records FROM fintech_finpay.gold.dim_channel
UNION ALL
SELECT 'dim_date' AS object_name, COUNT(*) AS records FROM fintech_finpay.gold.dim_date
UNION ALL
SELECT 'fact_transactions' AS object_name, COUNT(*) AS records FROM fintech_finpay.gold.fact_transactions;
```
